```python
from sklearn import manifold, datasets
import tensorflow.keras.backend as tf
n_samples = 10000
x, color = datasets.make_s_curve(n_samples, random_state=0)

from numpy import array
from keras.models import Sequential
from keras.layers import Dense
print(x.shape)
batch_size=500
lamda=0.001
```

```python
model = Sequential()
model.add((Dense(3,input_shape=(3,),activation="relu")))
model.add((Dense(3,activation="relu")))
model.add((Dense(2,activation="relu",name="extracted")))
model.add((Dense(3,activation="relu")))
model.add((Dense(3,activation="relu")))
```

```python

output = tf.function(model.input,model.get_layer('extracted').output)
```

```python
import math
def ecludianDistance(xi,xj):    
    s = tf.sum(
                        tf.square(
                                    tf.subtract(xi,xj)
                                )
                    )        
    return tf.sqrt(s)

def sammonError(y,x,batch_size):
    s = tf.sum(
                        tf.stack(
                            [
                                tf.divide((ecludianDistance(y[i],y[j]) - ecludianDistance(x[i],x[j]))**2,ecludianDistance(y[i],y[j])) for i in range(batch_size) for j in range(i+1, batch_size)
                            ]
                        )
                    )
    c = tf.sum(
                        tf.stack(
                            [
                                ecludianDistance(y[i],y[j]) for i in range(batch_size) for j in range(i+1, batch_size)
                            ]
                        )
                     )
    return tf.sum(tf.divide(s,c))
    

def loss(y_actual,y_predicted):
    
    
    return tf.mean(tf.square(y_actual-y_predicted)) + lamda*sammonError(y_actual,output(y_actual),batch_size)
```

```python
model.compile(optimizer = 'adam', loss = loss)
model.summary()
```

```python
model.fit(x,x, epochs=100, batch_size=batch_size, verbose=1, validation_split=0.2)
```
